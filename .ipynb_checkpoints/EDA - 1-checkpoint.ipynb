Eliminación de datos nulos (menor al 5%)

```python
limite = len(df) * 0.05
print(limite)
```

```python
columnas_para_eliminar = df.columns[df.isna().sum() <= limite]
print(columnas_para_eliminar)
```

```python
df_limpio = df.dropna(subset=columnas_para_eliminar).copy()
print(df_limpio.isna().sum().sort_values(ascending=False))
```

```python
sns.displot(data=df_limpio, x="Employee_Location", kind="hist")
plt.xticks(rotation=90)
plt.show()
```

```python
columnas_restantes = df_limpio.columns[df_limpio.isna().sum() > 0]
print(columnas_restantes)
for columna in columnas_restantes:
    moda = df_limpio.loc[:, columna].mode()[0]
    df_limpio.loc[:, columna].fillna(moda, inplace=True)
```

```python
print(df_limpio.isna().sum())
```

```python
for columna in listaColumnas:
    print(columna, df[columna].nunique())
```

```python
display(df_limpio["Designation"].head(20))
display(df_limpio["Designation"].unique())
```

```python
categorias = ["cientifico", "analista", "ingeniero", "ml", "consultor"]
cientifico = "Data Scientist|Science|Scientist|Specialist"
analista = "Data Analyst|Analytics"
ingeniero = "Data Engineer|Engineering|Architect|ETL|Infrastructure"
ml_ingeniero = "Machine Learning|ML|Big Data|AI"
consultor = "Consultant|Freelance"

condiciones = [
    (df_limpio["Designation"].str.contains(cientifico)),
    (df_limpio["Designation"].str.contains(analista)),
    (df_limpio["Designation"].str.contains(ingeniero)),
    (df_limpio["Designation"].str.contains(ml_ingeniero)),
    (df_limpio["Designation"].str.contains(consultor))
]
```

```python
df_limpio["CategoriaTrabajo"] = np.select(condiciones, categorias, default="Otro")
```

```python
display(df_limpio.head(20))
```

```python
df_limpio["CategoriaTrabajo"] = df_limpio["CategoriaTrabajo"].astype("category")
print(df_limpio.info())
```

```python
df_dummies = pd.get_dummies(df_limpio["CategoriaTrabajo"])
print(df_dummies)
```

```python
df_limpio = pd.concat([df_limpio, df_dummies], axis=1)
display(df_limpio)
```

```python
sns.boxplot(data=df_limpio, x="Salary_USD")
```

```python
q3 = df_limpio["Salary_USD"].quantile(0.75)
q1 = df_limpio["Salary_USD"].quantile(0.25)

iqr = q3 - q1
```

```python
lsuperior = q3 + (1.5 * iqr)
linferior = q1 - (1.5 * iqr)
```

```python
df_limpio_total = df_limpio[(df_limpio["Salary_USD"] > linferior) & (df_limpio["Salary_USD"] < lsuperior)].copy()
```

```python
sns.displot(data=df_limpio_total, x="Salary_USD")
```

```python
sns.boxplot(data=df_limpio_total, x="Salary_USD")
```
